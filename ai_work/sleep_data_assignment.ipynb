```python
#grup the above dataframe(df) by occupation,,highlight the column of sleep duration(take note of the way the column name is written) find the mean of the duration and sort it(auto in ascending order)
#call the first index wic has the lowest(index 0)
occ_dur=df.groupby('Occupation')['Sleep Duration'].mean().sort_values()
lowest_sleep_occ=occ_dur.index[0]
print(lowest_sleep_occ)
#print(occ_dur)
```

```python
bmi_insomnia_ratios={}
for bmi,group in df.groupby('BMI Category'):
    insomnia_count=len(group[group['Sleep Disorder']=='Insomnia'])
    total_count=len(group)
    ratio=insomnia_count/total_count
    bmi_insomnia_ratios[bmi]=round(ratio,2)
print(f"bmi_insomnia_ratios:{bmi_insomnia_ratios}")
#print(insomnia_count)
#print(total_count)
#print(bmi)
```
